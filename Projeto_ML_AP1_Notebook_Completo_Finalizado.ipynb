## üîç 3. Teste de Normalidade

### Justificativa:
Testar a normalidade das vari√°veis num√©ricas √© importante para validar a aplica√ß√£o de modelos lineares, que assumem essa distribui√ß√£o como uma de suas premissas.

Estat√≠stica W: Varia entre 0 e 1; valores pr√≥ximos de 1 indicam maior ader√™ncia √† normalidade.

Valor p: Um valor p muito pequeno (geralmente menor que 0,05) sugere evid√™ncias contra a hip√≥tese nula de normalidade.

```r
# Shapiro-Wilk para BMI
resultado_bmi <- shapiro.test(df$BMI)
print(resultado_bmi)

# Shapiro-Wilk para frequ√™ncia de treino
resultado_freq <- shapiro.test(df$Workout_Frequency..days.week.)
print(resultado_freq)
```

### Interpreta√ß√£o:
- `BMI`: valor p ‚âà 1,44 √ó 10‚Åª¬π‚Å¥, o que indica forte rejei√ß√£o da normalidade.
- `Workout_Frequency`: p < 2.2 √ó 10‚Åª¬π‚Å∂, tamb√©m indicando distribui√ß√£o n√£o normal.

Conclus√£o: recomenda-se o uso de m√©tricas n√£o param√©tricas ou transforma√ß√£o dos dados.

## üîó 4. Correla√ß√£o

### Justificativa:
A an√°lise de correla√ß√£o identifica rela√ß√µes entre vari√°veis que podem ser exploradas na regress√£o. Spearman foi escolhido por n√£o exigir normalidade.

* **Coeficiente rho ($œÅ$)**: Este coeficiente varia de -1 a 1 e mede a for√ßa e a dire√ß√£o da associa√ß√£o monot√¥nica entre duas vari√°veis. Um valor pr√≥ximo de 0 indica uma associa√ß√£o muito fraca ou inexistente.

* **Valor p**: Este valor indica a probabilidade de observar uma correla√ß√£o t√£o extrema quanto a encontrada, assumindo que a hip√≥tese nula (de que n√£o h√° correla√ß√£o) seja verdadeira. Um valor p alto (geralmente maior que 0,05) sugere que n√£o h√° evid√™ncias estat√≠sticas suficientes para rejeitar a hip√≥tese nula.

```r
# Correla√ß√£o de Spearman
cor.test(df$Workout_Frequency..days.week., df$Age, method = "spearman", exact = FALSE)
```

## ‚úé 5. Regress√£o Linear Simples

### Justificativa:
Utilizada para estimar uma vari√°vel cont√≠nua (`Calories_Burned`) a partir de uma preditora (`Workout_Frequency`). Permite entender e prever o gasto cal√≥rico.

```r
# Treinamento e avalia√ß√£o do modelo linear
set.seed(123)
indices_treino <- sample(seq_len(nrow(df)), size = 0.7 * nrow(df))
treino <- df[indices_treino, ]
teste <- df[-indices_treino, ]

modelo <- lm(Calories_Burned ~ Workout_Frequency..days.week., data = treino)
summary(modelo)

predicoes <- predict(modelo, newdata = teste)
mae <- mean(abs(predicoes - teste$Calories_Burned))
rmse <- sqrt(mean((predicoes - teste$Calories_Burned)^2))
sst <- sum((teste$Calories_Burned - mean(teste$Calories_Burned))^2)
ssr <- sum((predicoes - mean(teste$Calories_Burned))^2)
r2 <- ssr / sst

cat("MAE:", mae, "\n")
cat("RMSE:", rmse, "\n")
cat("R¬≤:", r2, "\n")
```

### Interpreta√ß√£o:
- **Erro M√©dio Absoluto (MAE) = 175,34**: erro m√©dio absoluto nas previs√µes.
    - Neste caso, indica que, em m√©dia, as previs√µes do modelo est√£o desviando dos valores reais em cerca de 175,34 unidades.
- **Raiz do Erro Quadr√°tico M√©dio (RMSE) = 215,44**: penaliza erros maiores.
    - Neste caso sugere que o desvio padr√£o das diferen√ßas entre os valores previstos e os reais √© de cerca de 215,44 unidades. O RMSE √© mais sens√≠vel a grandes erros do que o MAE, pois penaliza mais fortemente os desvios maiores.
- **RCoeficiente de Determina√ß√£o (R¬≤) = 0,343**: 34,3% da variabilidade das calorias √© explicada pela frequ√™ncia.
    - Neste caso sugere que cerca de 34,3% da variabilidade nos dados √© explicada pelo modelo, enquanto os 65,7% restantes s√£o atribu√≠dos a fatores n√£o explicados pelo modelo.


### Conclus√µes:
* **Precis√£o do Modelo**: Os valores de MAE e RMSE indicam que o modelo apresenta um erro m√©dio consider√°vel nas previs√µes. A magnitude desses erros sugere que as previs√µes podem n√£o ser suficientemente precisas para aplica√ß√µes que exigem alta exatid√£o.

* **Capacidade Explicativa**: O R¬≤ de 0,343 indica que o modelo explica uma parte limitada da variabilidade dos dados. Em termos pr√°ticos, isso significa que h√° outros fatores ou vari√°veis que influenciam significativamente a vari√°vel resposta e que n√£o est√£o sendo considerados pelo modelo atual.
* **Sensibilidade a Grandes Erros**: A diferen√ßa entre o MAE e o RMSE sugere a presen√ßa de alguns erros maiores, j√° que o RMSE penaliza mais fortemente grandes desvios. Isso indica que o modelo pode estar cometendo erros substanciais em certas previs√µes.


## üî¢ 6. Regress√£o Log√≠stica

### Justificativa:
O objetivo √© prever o g√™nero de uma pessoa com base em sua altura. A regress√£o log√≠stica permite classificar observa√ß√µes com sa√≠da bin√°ria.

```r
# Ajuste do modelo log√≠stico e avalia√ß√£o
sum(is.na(df$Gender))
sum(is.na(df$Height..m.))
df$Gender <- ifelse(df$Gender == "Male", 1, 0)

set.seed(123)
indice <- sample(1:nrow(df), size = 0.7 * nrow(df))
treinamento <- df[indice, ]
teste <- df[-indice, ]

modelo <- glm(Gender ~ Height..m., data = treinamento, family = binomial)
probabilidades <- predict(modelo, newdata = teste, type = "response")
limiar <- 0.5
previsoes <- ifelse(probabilidades > limiar, 1, 0)

install.packages("caret", dependencies = TRUE)
library(caret)
previsoes <- factor(previsoes, levels = c(0, 1))
valores_reais <- factor(teste$Gender, levels = c(0, 1))
matriz_confusao <- confusionMatrix(previsoes, valores_reais)
print(matriz_confusao)
```

### Interpreta√ß√£o:
- **Acur√°cia = 72,26%**: Propor√ß√£o total de previs√µes corretas
    - Bom resultado para uma vari√°vel simples.
- **Sensibilidade e Especificidade ‚âà 73%**: balan√ßo entre classes.
    - O modelo acertou aproximadamente 73% das previs√µes.‚Äã
- **Kappa = 0,445**: concord√¢ncia moderada.
    - Indica uma concord√¢ncia moderada entre as previs√µes do modelo e os valores reais.
- **Teste de McNemar**: Avalia se h√° diferen√ßas significativas entre as taxas de erro do modelo nas duas classes.
    - **P-Value** : 0,6567: Um p-valor alto sugere que n√£o h√° diferen√ßa significativa entre as taxas de erro, indicando que o modelo n√£o favorece uma classe em detrimento da outra.

Conclus√£o:

O modelo apresenta uma acur√°cia razo√°vel de 72,26%, com sensibilidades e especificidades equilibradas em torno de 73%. No entanto, o valor de Kappa indica uma concord√¢ncia moderada, sugerindo que h√° espa√ßo para melhorias.

O modelo consegue capturar bem a diferen√ßa entre g√™neros com base na altura, apesar da simplicidade.
