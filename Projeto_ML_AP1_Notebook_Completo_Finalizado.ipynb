## 🔍 3. Teste de Normalidade

### Justificativa:
Testar a normalidade das variáveis numéricas é importante para validar a aplicação de modelos lineares, que assumem essa distribuição como uma de suas premissas.

Estatística W: Varia entre 0 e 1; valores próximos de 1 indicam maior aderência à normalidade.

Valor p: Um valor p muito pequeno (geralmente menor que 0,05) sugere evidências contra a hipótese nula de normalidade.

```r
# Shapiro-Wilk para BMI
resultado_bmi <- shapiro.test(df$BMI)
print(resultado_bmi)

# Shapiro-Wilk para frequência de treino
resultado_freq <- shapiro.test(df$Workout_Frequency..days.week.)
print(resultado_freq)
```

### Interpretação:
- `BMI`: valor p ≈ 1,44 × 10⁻¹⁴, o que indica forte rejeição da normalidade.
- `Workout_Frequency`: p < 2.2 × 10⁻¹⁶, também indicando distribuição não normal.

Conclusão: recomenda-se o uso de métricas não paramétricas ou transformação dos dados.

## 🔗 4. Correlação

### Justificativa:
A análise de correlação identifica relações entre variáveis que podem ser exploradas na regressão. Spearman foi escolhido por não exigir normalidade.

* **Coeficiente rho ($ρ$)**: Este coeficiente varia de -1 a 1 e mede a força e a direção da associação monotônica entre duas variáveis. Um valor próximo de 0 indica uma associação muito fraca ou inexistente.

* **Valor p**: Este valor indica a probabilidade de observar uma correlação tão extrema quanto a encontrada, assumindo que a hipótese nula (de que não há correlação) seja verdadeira. Um valor p alto (geralmente maior que 0,05) sugere que não há evidências estatísticas suficientes para rejeitar a hipótese nula.

```r
# Correlação de Spearman
cor.test(df$Workout_Frequency..days.week., df$Age, method = "spearman", exact = FALSE)
```

### Interpretação:
- Correlação próxima de zero (0.0063) indica independência entre variáveis.
- Valor p elevado (> 0.8) confirma que não há evidência estatística de associação.

### Conclusão:

As variações na frequência semanal de exercícios não parecem estar associadas de forma monotônica com a idade dos indivíduos na sua amostra.​

## ✎ 5. Regressão Linear Simples

### Justificativa:
Utilizada para estimar uma variável contínua (`Calories_Burned`) a partir de uma preditora (`Workout_Frequency`). Permite entender e prever o gasto calórico.

```r
# Treinamento e avaliação do modelo linear
set.seed(123)
indices_treino <- sample(seq_len(nrow(df)), size = 0.7 * nrow(df))
treino <- df[indices_treino, ]
teste <- df[-indices_treino, ]

modelo <- lm(Calories_Burned ~ Workout_Frequency..days.week., data = treino)
summary(modelo)

predicoes <- predict(modelo, newdata = teste)
mae <- mean(abs(predicoes - teste$Calories_Burned))
rmse <- sqrt(mean((predicoes - teste$Calories_Burned)^2))
sst <- sum((teste$Calories_Burned - mean(teste$Calories_Burned))^2)
ssr <- sum((predicoes - mean(teste$Calories_Burned))^2)
r2 <- ssr / sst

cat("MAE:", mae, "\n")
cat("RMSE:", rmse, "\n")
cat("R²:", r2, "\n")
```

### Interpretação:
- **Erro Médio Absoluto (MAE) = 175,34**: erro médio absoluto nas previsões.
    - Neste caso, indica que, em média, as previsões do modelo estão desviando dos valores reais em cerca de 175,34 unidades.
- **Raiz do Erro Quadrático Médio (RMSE) = 215,44**: penaliza erros maiores.
    - Neste caso sugere que o desvio padrão das diferenças entre os valores previstos e os reais é de cerca de 215,44 unidades. O RMSE é mais sensível a grandes erros do que o MAE, pois penaliza mais fortemente os desvios maiores.
- **RCoeficiente de Determinação (R²) = 0,343**: 34,3% da variabilidade das calorias é explicada pela frequência.
    - Neste caso sugere que cerca de 34,3% da variabilidade nos dados é explicada pelo modelo, enquanto os 65,7% restantes são atribuídos a fatores não explicados pelo modelo.


### Conclusões:
* **Precisão do Modelo**: Os valores de MAE e RMSE indicam que o modelo apresenta um erro médio considerável nas previsões. A magnitude desses erros sugere que as previsões podem não ser suficientemente precisas para aplicações que exigem alta exatidão.

* **Capacidade Explicativa**: O R² de 0,343 indica que o modelo explica uma parte limitada da variabilidade dos dados. Em termos práticos, isso significa que há outros fatores ou variáveis que influenciam significativamente a variável resposta e que não estão sendo considerados pelo modelo atual.
* **Sensibilidade a Grandes Erros**: A diferença entre o MAE e o RMSE sugere a presença de alguns erros maiores, já que o RMSE penaliza mais fortemente grandes desvios. Isso indica que o modelo pode estar cometendo erros substanciais em certas previsões.


## 🔢 6. Regressão Logística

### Justificativa:
O objetivo é prever o gênero de uma pessoa com base em sua altura. A regressão logística permite classificar observações com saída binária.

```r
# Ajuste do modelo logístico e avaliação
sum(is.na(df$Gender))
sum(is.na(df$Height..m.))
df$Gender <- ifelse(df$Gender == "Male", 1, 0)

set.seed(123)
indice <- sample(1:nrow(df), size = 0.7 * nrow(df))
treinamento <- df[indice, ]
teste <- df[-indice, ]

modelo <- glm(Gender ~ Height..m., data = treinamento, family = binomial)
probabilidades <- predict(modelo, newdata = teste, type = "response")
limiar <- 0.5
previsoes <- ifelse(probabilidades > limiar, 1, 0)

install.packages("caret", dependencies = TRUE)
library(caret)
previsoes <- factor(previsoes, levels = c(0, 1))
valores_reais <- factor(teste$Gender, levels = c(0, 1))
matriz_confusao <- confusionMatrix(previsoes, valores_reais)
print(matriz_confusao)
```

### Interpretação:
- **Acurácia = 72,26%**: Proporção total de previsões corretas
    - Bom resultado para uma variável simples.
- **Sensibilidade e Especificidade ≈ 73%**: balanço entre classes.
    - O modelo acertou aproximadamente 73% das previsões.​
- **Kappa = 0,445**: concordância moderada.
    - Indica uma concordância moderada entre as previsões do modelo e os valores reais.
- **Teste de McNemar**: Avalia se há diferenças significativas entre as taxas de erro do modelo nas duas classes.
    - **P-Value** : 0,6567: Um p-valor alto sugere que não há diferença significativa entre as taxas de erro, indicando que o modelo não favorece uma classe em detrimento da outra.

Conclusão:

O modelo apresenta uma acurácia razoável de 72,26%, com sensibilidades e especificidades equilibradas em torno de 73%. No entanto, o valor de Kappa indica uma concordância moderada, sugerindo que há espaço para melhorias.

O modelo consegue capturar bem a diferença entre gêneros com base na altura, apesar da simplicidade.
